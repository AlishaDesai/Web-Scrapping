# Flipkart customer reviews and ratings

```python
import pandas as pd
```

### function to find urls on one page

```python
def findurls_inapage(baseurl):
    from bs4 import BeautifulSoup
    import requests
    url = baseurl
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    urlsinapage = []
    for i in soup.find_all("div",class_={"_1YokD2 _3Mn1Gg"}):
        for j in i.find_all("div",class_={"_1AtVbE col-12-12"}):
            for k in j.find_all("div",class_={"_13oc-S"}):
                for l in k.find_all("a"):
                    link = l['href']
                    full_url="https://www.flipkart.com"+(link)
                    urlsinapage.append(full_url)
    return urlsinapage
```

### function to find overall ratings and review

```python
def ratings_n_reviews(prod_url):
    from bs4 import BeautifulSoup
    import requests
    name = []
    overall_ratings = []
    total_no_ratings = []
    total_no_review = []
    for url in prod_url:
        data = requests.get(url)
        soup = BeautifulSoup(data.content,'html.parser')
        for i in soup.find_all("h1",class_=["yhB1nd"]):
            name.append(i.text)
        for i in soup.find_all("div",class_=["_2d4LTz"]):
            overall_ratings.append(i.text)
        w = []
        u=[]
        x=0
        for i in soup.find_all("div",class_=["row _2afbiS"]):
            if x==0:
                w.append(i.text)
            else:
                u.append(i.text)
            x=x+1

        if len(w) >= 1:
            total_no_ratings.append(w[len(w)-1])
        else:
            total_no_ratings.append("unavailable")
        if len(u) >= 1:
            total_no_review.append(u[len(u)-1])
        else:
            total_no_review.append("unavailable")
    return name,overall_ratings,total_no_ratings,total_no_review
```

### function to find detailed ratings

```python
def get_rating_split(prod_url):
    from bs4 import BeautifulSoup
    import requests
    five = []
    four = []
    three = []
    two = []
    one = []
    for url in prod_url:
        data = requests.get(url)
        soup = BeautifulSoup(data.content,'html.parser')
        a=[]
        b = []
        c=[]
        d=[]
        e=[]
        x=0
        for i in soup.find_all("div",class_=["_1uJVNT"]):
            if x==0:
                a.append(i.text)
            elif x==1:
                b.append(i.text)
            elif x==2:
                c.append(i.text)
            elif x==3:
                d.append(i.text)
            else:
                e.append(i.text)
            x=x+1

        if len(a) >= 1:
            five.append(a[len(a)-1])
        else:
            five.append("unavailable")
        if len(b) >= 1:
            four.append(b[len(b)-1])
        else:
            four.append("unavailable")
        if len(c) >= 1:
            three.append(c[len(c)-1])
        else:
            three.append("unavailable")
        if len(d) >= 1:
            two.append(d[len(d)-1])
        else:
            two.append("unavailable")
        if len(e) >= 1:
            one.append(e[len(e)-1])
        else:
            one.append("unavailable")
    return five,four,three,two,one
```

### function to find detail reviews

```python
def get_detail_review(prod_url):
    from bs4 import BeautifulSoup
    import requests
    review1=[]
    for url in prod_url:
        data = requests.get(url)
        soup = BeautifulSoup(data.content,'html.parser')

        r1=""
        for i in soup.find_all("div",class_=["t-ZTKy"]):
            r1 = r1 + str(i.text)

        if len(r1) >= 1:
            review1.append(r1)
        else:
            review1.append("review unavailable")
    return review1
```

### generate urls

```python
baseurls = []
allurls = []

for i in range(1,25,1):
    url = "https://www.flipkart.com/search?q=tv+smart+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=tv+smart+tv%7CTelevisions&requestId=0eabcd31-e933-4ed7-8172-c8ffcd640436&as-searchtext=tv&page="
    url = url + str(i)
    baseurls.append(url)
    
for i in baseurls:
    allurls.extend(findurls_inapage(i))
```

### extract data

```python
name=[]
overall_ratings=[]
tota_no_ratings=[]
total_no_review=[]
name,overall_ratings,total_no_ratings,total_no_review=ratings_n_reviews(allurls)
```

```python
one=[]
two=[]
three=[]
four=[]
five=[]
five,four,three,two,one=get_rating_split(allurls)
```

```python
review1=[]

review1=get_detail_review(allurls)
```

### Make Data Frame

```python
Q = pd.DataFrame([name,overall_ratings,tota_no_ratings,total_no_review,five,four,three,two,one,review1]).T
Q.columns = ["Name","overall_ratings","tota_no_ratings","total_no_review","five-star","four-star","three","two-star","one-star","reviews"]
```
